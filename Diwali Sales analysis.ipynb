#EDA(Exploratory Data Analysis)




```python
sns.countplot(x='Gender', data=df)
bx = sns.countplot(x='Gender', data=df)
for bars in bx.containers:
  bx.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_state = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_state, x = 'Gender',y= 'Amount')
```

From above grapgh we noticed that we have more women buyers and even female purchaing power is also more than men

```python
ax = sns.countplot(x='Age Group', data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Age Group', data=df, hue='Gender')

#ax = sns.countplot(x='Age Group', data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_age, x = 'Age Group',y= 'Amount')
```

Here we can noticed more buyers are from the age group of 26-35
